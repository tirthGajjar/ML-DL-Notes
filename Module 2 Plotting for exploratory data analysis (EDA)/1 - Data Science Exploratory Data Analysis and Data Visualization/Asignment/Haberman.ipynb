# EDA On Haberman's Survival Data Set

### Description

The dataset contains cases from a study that was conducted between 1958 and 1970 at the University of Chicago's Billings Hospital on the survival of patients who had undergone surgery for breast cancer.

### Data & Columns
- Number of Instances: 306
- Number of Attributes: 4 (including the class attribute)
    1.  Age of patient at time of operation (numerical)
    2. Patient's year of operation (year - 1900, numerical)
    3. Number of positive axillary nodes detected (numerical)
    4. Survival status (class attribute) 1 = the patient survived 5 years or longer 2 = the patient died within 5 year
- Missing Attribute Values: None

## Univariate Analysis

- Univariate analysis explores each variable in a data set, separately. It looks at the range of values, as well as the central tendency of the values. It describes the pattern of response to the variable. It describes each variable on its own.
- Descriptive statistics describe and summarize data. Univariate descriptive statistics describe individual variables.

Method to perform univariate analysis will depend on whether the variable type is categorical or continuous

In this dataset,

## Let's check the distribution of the data based on the survival status of the patient

```python
f, (boxplots_ax, distplot_ax) = plt.subplots(2, 3, figsize=(18, 3), sharey= 'col', gridspec_kw={"height_ratios": (.5, .5)})

for ax in boxplots_ax:
    ax.tick_params(top=False, bottom=False, left=False, right=False, labelleft=True, labelbottom=False)

sns.boxplot( df[df['status']==1]["age"] , color=current_palette[0], ax=boxplots_ax[0]).set(xlabel='',ylabel='survived')
sns.boxplot( df[df['status']==1]["year"] , color=current_palette[0], ax=boxplots_ax[1]).set(xlabel='')
sns.boxplot( df[df['status']==1]["nodes"] , color=current_palette[0], ax=boxplots_ax[2]).set(xlabel='')
sns.boxplot( df[df['status']==2]["age"] , color=current_palette[1], ax=distplot_ax[0]).set(ylabel='died')
sns.boxplot( df[df['status']==2]["year"] , color=current_palette[1], ax=distplot_ax[1])
sns.boxplot( df[df['status']==2]["nodes"] , color=current_palette[1], ax=distplot_ax[2])
```

A boxplot can summarize the distribution of a numeric variable for several groups. 
The problem here is that summarizing also means losing information, and that can be a pitfall. 

In the boxplot above, <span class="girk">it is easy to conclude that patients with higher number of nodes has a higher probablity of not surviving. 
However, we cannot see the underlying distribution of dots in each group or their number of observations</span>

The amount of data we are working with is not too large, adding jitter on top of your boxplot can make the chart more insightful

```python
current_palette = sns.color_palette("muted").as_hex();

f, (boxplots_ax, distplot_ax) = plt.subplots(2, 3, figsize=(18, 3), sharey= 'col', gridspec_kw={"height_ratios": (.5, .5)})

for ax in boxplots_ax:
    ax.tick_params(top=False, bottom=False, left=False, right=False, labelleft=True, labelbottom=False)

sns.boxplot( df[df['status']==1]["age"] , color=current_palette[0], ax=boxplots_ax[0]).set(xlabel='',ylabel='survived')
sns.boxplot( df[df['status']==1]["year"] , color=current_palette[0], ax=boxplots_ax[1]).set(xlabel='')
sns.boxplot( df[df['status']==1]["nodes"] , color=current_palette[0], ax=boxplots_ax[2]).set(xlabel='')

sns.stripplot( df[df['status']==1]["age"] , color=current_palette[1], jitter=0.5, size=2.5, ax=boxplots_ax[0]).set(xlabel='')
sns.stripplot( df[df['status']==1]["year"] , color=current_palette[1], jitter=0.5, size=2.5, ax=boxplots_ax[1]).set(xlabel='')
sns.stripplot( df[df['status']==1]["nodes"] , color=current_palette[1], jitter=0.5, size=2.5, ax=boxplots_ax[2]).set(xlabel='')

sns.boxplot( df[df['status']==2]["age"] , color=current_palette[1], ax=distplot_ax[0]).set(ylabel='died')
sns.boxplot( df[df['status']==2]["year"] , color=current_palette[1], ax=distplot_ax[1])
sns.boxplot( df[df['status']==2]["nodes"] , color=current_palette[1], ax=distplot_ax[2])

sns.stripplot( df[df['status']==2]["age"] , color=current_palette[4], jitter=0.5, size=2.5, ax=distplot_ax[0])
sns.stripplot( df[df['status']==2]["year"] , color=current_palette[4], jitter=0.5, size=2.5, ax=distplot_ax[1])
sns.stripplot( df[df['status']==2]["nodes"] , color=current_palette[4], jitter=0.5, size=2.5, ax=distplot_ax[2])
```

## Bi-variate Analysis
Bi-variate Analysis finds out the relationship between two variables.   
Here, we look for association and disassociation between variables at a pre-defined significance level.   
We can perform bi-variate analysis for any combination of categorical and continuous variables.   
The combination can be: 
1. Categorical & Categorical, 
2. Categorical & Continuous and 
3. Continuous & Continuous.   

Different methods are used to tackle these combinations during analysis process.

```python
 sns.pairplot(df, hue="status_name")
```

Just to be sure that we are not missing any corelation in ur observation, let's plot corelation in a heatmap way

```python
# let's first revert the status column to numeric from category

cat_columns = df.select_dtypes(['category']).columns
df[cat_columns] = df[cat_columns].apply(lambda x: x.cat.codes)
sns.set(style="white")

df2 = df.copy(deep=True)
df2['nodes'] = df2.apply(lambda row: 0 if row['nodes'] <5 else 1, axis=1)

# Compute the correlation matrix
corr = df.corr()
corr2 = df2.corr()
# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(1,2,figsize=(15, 10))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 20, sep=20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot=True,ax=ax[0])

sns.heatmap(corr2, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot=True,ax=ax[1])
```

As one can assume, there shouldn't be any corelation between age, year of count of nodes columns.  
From the above pairplot, we can deduce the same that there is no corelation between any of independent variables of the dataset 

From the pairplot, we can see that Number of positive axillary nodes has positive corelation with survival status

```python
from collections import defaultdict
from random import shuffle

def draw_chart(data,groupby_category='nodes', target_category='status', sns_color_palette ="Paired"):
    current_palette = sns.color_palette(sns_color_palette).as_hex();
    # plot
    barWidth = 0.25
    fig = plt.figure(figsize=(15,4))
    ax = fig.add_subplot(111)
    
    r = []
    d = defaultdict(list)
    category_unique_values = data[target_category].unique();
    for name,group in data.groupby(groupby_category):
        total = group[target_category].count();
        r.append(name)
        for item in category_unique_values:
            d[item].append(
                    group[group[target_category]==item][target_category].count()/total*100
            )
    
    r_index = r
    for index, (key, values) in enumerate(d.items()):
        plt.bar(r_index, values, color=current_palette[index % len(current_palette)], width=barWidth, edgecolor='white', label=key)
        r_index = np.array(r_index) + barWidth
    
    
    plt.xlabel(groupby_category)
    plt.ylabel('percentage share')
    plt.xticks([r2 + barWidth/len(category_unique_values) for r2 in r],r)
    plt.legend()

     # Show graphic
    plt.show()
```

```python
from collections import defaultdict
from random import shuffle

def autolabel(bar_plot):
    for idx,rect in enumerate(bar_plot):
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., (rect.get_y()+ 0.4*rect.get_height()),
                round(height, 2),
                bbox=dict(facecolor='white', alpha=0.5),
                ha='center', va='bottom')

def draw_chart(data,groupby_category='nodes', target_category='status', sns_color_palette ="Paired", figsize=(15,4)):
    current_palette = sns.color_palette(sns_color_palette).as_hex();
    # plot
    barWidth = 0.85
    fig = plt.figure(figsize=figsize)
    ax = fig.add_subplot(111)
    
    r = []
    d = defaultdict(list)
    category_unique_values = data[target_category].unique();
    for name,group in data.groupby(groupby_category):
        total = group[target_category].count();
        r.append(name)
        for item in category_unique_values:
            d[item].append(
                    group[group[target_category]==item][target_category].count()/total*100
            )
    
    current_total=np.zeros((len(r)))
    for index, (key, values) in enumerate(d.items()):
        bar_plt = plt.bar(r, values, bottom=current_total, color=current_palette[index % len(current_palette)], width=barWidth, edgecolor='white', label=key)
        current_total = current_total + values
        
    autolabel(bar_plt)
    plt.xlabel(groupby_category)
    plt.ylabel('percentage share')
    plt.legend()
    plt.xticks(r)

     # Show graphic
    plt.show()
```

```python
df_m['status_name'] = df_m.apply(lambda row: 'survived' if row.status == 1 else 'not survived', axis=1)
draw_chart(groupby_category='age',target_category='status_name', data=df_m, sns_color_palette='Set2')
```
